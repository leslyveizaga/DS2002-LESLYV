```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import kagglehub
import csv
import seaborn as sns
```

```python
df = pd.read_csv('nbaplayersdraft.csv')

print(df.head())    # First 5 rows
print(df.columns)   # Column names
```

```python
#Matplotlib
plt.hist(df['years_active'], bins=15, color='blue', edgecolor='black')
plt.title('Distribution of Player Career Length')
plt.xlabel('Years Active')
plt.ylabel('Number of Players')
plt.show()
```

```python
#Iris dataset
iris = pd.read_csv('nbaplayersdraft.csv')
print(iris.head())

#Seaborn
numeric_columns = ['years_active', 'win_shares', 'points_per_game', 'games']
sns.pairplot(iris[numeric_columns])
plt.title('Pairplot of NBA Players Draft Data')
plt.show()

```
