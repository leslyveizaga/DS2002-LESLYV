```python
# - Group the players by School and calculate the total points scored by each school.
# team_avg_stats_regular = regular_season_df.groupby('Team')[['Points', 'Rebounds', 'Assists']].mean().reset_index()

# Total_points_by_school =
accplayers2324.groupby('School')['PTS'].sum().reset_index()
```

```python
#- Group the players by School and calculate the total assists (AST) for each team.

accplayers2324.groupby('School')['AST'].sum().reset_index()
```

```python
#- Sort the schools by total points scored and display the top 3 schools.
accplayers2324.groupby('School')['PTS'].sum().sort_values(ascending=False).head(3)
```

```python
# Create a bar chart showing the top 5 players by total points scored (PTS).

top5pts = accplayers2324[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(5,6))
sns.barplot(x='PTS', y='Player', data=top5pts)
plt.title('Top 5 Players by PTS')
plt.show()
```

```python
# - Correlation: Investigate whether there is any correlation between a player's field goal
# percentage (FG%) and their win shares (WS). Provide a scatter plot and calculate the
# correlation coefficient.
fg = accplayers2324['FG%']
ws = accplayers2324['WS']
plt.title ('FG% vs. WS scatter plot')
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')

plt.scatter(x=fg, y=ws)
plt.show()

print('Correlation coeff:')
print(fg.corr(ws))
```
